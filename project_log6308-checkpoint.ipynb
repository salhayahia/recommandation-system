## Quelques exemples de titres en double ne diffèrent que par les derniers mots, on va donc les supprimer

```python
indices = []
for i,row in data_sorted.iterrows():
    indices.append(i)
```

```python
import itertools
stage1_dedupe_asins = []
i = 0
j = 0
num_data_points = data_sorted.shape[0]
while i < num_data_points and j < num_data_points:
    
    previous_i = i
    a = data_sorted['title'].loc[indices[i]].split()

    # rechercher séquentiellement les produits similaires 
    j = i+1
    while j < num_data_points:
        b = data_sorted['title'].loc[indices[j]].split()

        # stocke la longueur maximale de deux chaînes
        length = max(len(a), len(b))

        # count est utilisé pour stocker le nombre de mots correspondants dans les deux chaînes
        count  = 0

        # itertools.zip_longest(a,b): mappera les mots correspondants dans les deux chaînes, il ajoutera None en cas de chaînes inégales
        for k in itertools.zip_longest(a,b): 
            if (k[0] == k[1]):
                count += 1
        #  si le nombre de mots dans lesquels les deux chaînes diffèrent sont> 2, nous l'examinons car ces deux attributs sont différents
        #  si le nombre de mots dans lesquels les deux chaînes diffèrent sont <2, nous les ignorons
        if (length - count) > 2: 
            #si les deux chaînes diffèrent de plus de 2 mots, nous incluons le premier index de chaîne
            stage1_dedupe_asins.append(data_sorted['asin'].loc[indices[i]])
            ## si la comparaison entre num_data_points, num_data_points-1 et qu'elles diffèrent par plus de 2 mots, nous incluons les deux
            if j == num_data_points-1: stage1_dedupe_asins.append(data_sorted['asin'].loc[indices[j]])
            # commence la recherche des produits similaires correspondants à la 2ème chaîne
            i = j
            break
        else:
            j += 1
    if previous_i == i:
        break
```

## Nous définissons certaines fonctions utilitaires que nous utiliserons plus tard:

```python
#fonction qui affiche une image
def display_img(url,ax,fig):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    plt.imshow(img)

#fonction qui trace les modèles pour que nous comprenions la décision de l'algorithme
def plot_heatmap(keys, values, labels, url, text):
        # keys: liste des mots du titre recommandé
        # valeurs: len (values) == len (keys), les valeurs (i) représentent l'occurrence du mot keys (i)
        # labels: len (labels) == len (keys), les valeurs des labels dépendent du modèle utilisé
        # url: URL du produit

        # nous allons diviser la figure entière en deux parties
        gs = gridspec.GridSpec(2, 2, width_ratios=[4,1], height_ratios=[4,1]) 
        fig = plt.figure(figsize=(25,3))
        
        # tracer la heat map qui répresente le nombre de mots fréquemment rencontrés dans title2
        ax = plt.subplot(gs[0])
        # il affiche une cellule de couleur blanche si le mot est intersection (liste de mots de title1 et liste de mots de title2),sinon en noir 
        ax = sns.heatmap(np.array([values]), annot=np.array([labels]))
        ax.set_xticklabels(keys)
        ax.set_title(text) 
        
        # afficher l'image du produit
        ax = plt.subplot(gs[1])
        ax.grid(False)
        ax.set_xticks([])
        ax.set_yticks([])
        
        display_img(url, ax, fig)
        plt.show()
    
def plot_heatmap_image(doc_id, vec1, vec2, url, text, model):
        # doc_id: index du title1
        # vec1: le vecteur d'input du produit, il est de type dict {word: count}
        # vec2: le vecteur des produits recommandés, il est de type dict {word: count}
        # url: URL de l'image du produit
        # text: titre du produit recommandé (utilisé pour conserver le titre de l'image)
        # modèle, se trouve parmi ces modèles là,
            # 1. bag_of_words
            # 2. tfidf
            # 3. idf    

    #nous trouvons les mots communs dans les deux titres, car seulement ces mots contribuent à la distance entre deux titres de titres
    intersection = set(vec1.keys()) & set(vec2.keys()) 

    # on met à zéro les valeurs des mots qui ne se croisent pas, c'est juste pour montrer la différence de heatmap
    for i in vec2:
        if i not in intersection:
            vec2[i]=0

    keys = list(vec2.keys())
    values = [vec2[x] for x in vec2.keys()]
    
    # labels: len (labels) == len (keys), les valeurs des labels dépendent du modèle utilisé
        # si model == 'bag of words': labels(i) = values(i)
        # si model == 'tfidf weighted bag of words':labels(i) = tfidf(keys(i))
        # si model == 'idf weighted bag of words':labels(i) = idf(keys(i))    

    if model == 'bag_of_words':
        labels = values
    elif model == 'tfidf':
        labels = []
        for x in vec2.keys():
            if x in  tfidf_title_vectorizer.vocabulary_:
                labels.append(tfidf_title_features[doc_id, tfidf_title_vectorizer.vocabulary_[x]])
            else:
                labels.append(0)
    elif model == 'idf':
        labels = []
        for x in vec2.keys():
            if x in  idf_title_vectorizer.vocabulary_:
                labels.append(idf_title_features[doc_id, idf_title_vectorizer.vocabulary_[x]])
            else:
                labels.append(0)

    plot_heatmap(keys, values, labels, url, text)

# cette fonction obtient une liste de mots avec la fréquence de chaque mot donné "text"
def text_to_vector(text):
    word = re.compile(r'\w+')
    words = word.findall(text)
    return Counter(words) # Counter compte l'occurrence de chaque mot de la liste, il retourne un objet de type dict {word: count}

def get_result(doc_id, content_a, content_b, url, model):
    text1 = content_a
    text2 = content_b
    # vector1 = dict{word11:#count, word12:#count, etc.}
    vector1 = text_to_vector(text1)
    # vector1 = dict{word21:#count, word22:#count, etc.}
    vector2 = text_to_vector(text2)
    plot_heatmap_image(doc_id, vector1, vector2, url, text2, model)     
```

# Bag of Words sur les titres des produits (BoW)

```python
from sklearn.feature_extraction.text import CountVectorizer
title_vectorizer = CountVectorizer()
title_features   = title_vectorizer.fit_transform(data['title'])
title_features.get_shape() # obtenir le nombre de lignes et de colonnes dans la matrice d'objets.
```

```python
def bag_of_words_model(doc_id, num_results):
    # doc_id: l'identifiant du produit dans le corpus 
    # pairwise_dist stockera la distance entre le produit d'entrée donné et tous les produits restants
    # la métrique utilisée ici est le cosinus, la distance du cosinus est mesurée par K(X, Y) = <X, Y> / (||X||*||Y||)
    pairwise_dist = pairwise_distances(title_features,title_features[doc_id])
    
    # np.argsort retournera des index des plus petites distances
    indices = np.argsort(pairwise_dist.flatten())[0:num_results]
    #pdists stokera les petites distances
    pdists  = np.sort(pairwise_dist.flatten())[0:num_results]

    #data frame indices des 9 plus petites distances
    df_indices = list(data.index[indices])
    
    for i in range(0,len(indices)):
        # nous allons afficher pass 1. doc_id, 2. title1, 3. title2, url, model
        get_result(indices[i],data['title'].loc[df_indices[0]], data['title'].loc[df_indices[i]], data['medium_image_url'].loc[df_indices[i]], 'bag_of_words')
        print('ASIN :',data['asin'].loc[df_indices[i]])
        print ('Brand:', data['brand'].loc[df_indices[i]])
        print ('Title:', data['title'].loc[df_indices[i]])
        print ('Euclidean similarity with the query image :', pdists[i])
        print('='*125)

#Appelez le modèle bag-of-words pour un produit  pour obtenir des produits similaires.
bag_of_words_model(12566, 10) 
# Dans la heat map affichée, chaque valeur représente la valeur de comptage
# pour le mot d'étiquette, la couleur représente l'intersection avec le titre entrées.
```

# TF-IDF 

```python
tfidf_title_vectorizer = TfidfVectorizer(min_df = 0)
tfidf_title_features = tfidf_title_vectorizer.fit_transform(data['title'])
```

```python
def tfidf_model(doc_id, num_results):
    # doc_id: l'identifiant du produit dans le corpus 
    # pairwise_dist stockera la distance entre le produit d'entrée donné et tous les produits restants
    # la métrique utilisée ici est le cosinus, la distance du cosinus est mesurée par K(X, Y) = <X, Y> / (||X||*||Y||)
    
    pairwise_dist = pairwise_distances(tfidf_title_features,tfidf_title_features[doc_id])

    # np.argsort retournera des index des plus petites distances
    indices = np.argsort(pairwise_dist.flatten())[0:num_results]
    #pdists stokera les petites distances
    pdists  = np.sort(pairwise_dist.flatten())[0:num_results]

    #data frame indices des 9 plus petites distances
    df_indices = list(data.index[indices])

    for i in range(0,len(indices)):
        # Nous allons afficher pass 1. doc_id, 2. title1, 3. title2, url, model
        get_result(indices[i], data['title'].loc[df_indices[0]], data['title'].loc[df_indices[i]], data['medium_image_url'].loc[df_indices[i]], 'tfidf')
        print('ASIN :',data['asin'].loc[df_indices[i]])
        print('BRAND :',data['brand'].loc[df_indices[i]])
        print ('Eucliden distance from the given image :', pdists[i])
        print('='*125)
tfidf_model(12566, 20)
# Dans la heat map affichée, chaque valeur représente les valeurs tfidf du mot d'étiquette
# la couleur représente l'intersection avec le titre d'entrées
```

# Essayons IDF seulement¶

### Nous essayons cette approche parce qu'en général un mot ne figure pas souvent dans un titre

```python
idf_title_vectorizer = CountVectorizer()
idf_title_features = idf_title_vectorizer.fit_transform(data['title'])
```

```python
def n_containing(word):
    # renvoie le nombre de documents contenant le mot donné
    return sum(1 for blob in data['title'] if word in blob.split())

def idf(word):
    # idf = log(#number de documents / #nombre de documents contenant le mot donné)
    return math.log(data.shape[0] / (n_containing(word)))
```

```python
# converir les valeurs en float
idf_title_features  = idf_title_features.astype(np.float)

for i in idf_title_vectorizer.vocabulary_.keys():
    # pour chaque mot du corpus entier nous allons trouver sa valeur idf
    idf_val = idf(i)
    
    # pour calculer idf_title_features, nous devons remplacer les valeurs de comptage par les valeurs idf du mot
    for j in idf_title_features[:, idf_title_vectorizer.vocabulary_[i]].nonzero()[0]:
        
        # nous remplaçons les valeurs de comptage du mot i dans le document j par idf_value du mot i
        idf_title_features[j,idf_title_vectorizer.vocabulary_[i]] = idf_val
```

```python
def idf_model(doc_id, num_results):
    # doc_id: l'identifiant du produit dans le corpus 
    # pairwise_dist stockera la distance entre le produit d'entrée donné et tous les produits restants
    # la métrique utilisée ici est le cosinus, la distance du cosinus est mesurée par K(X, Y) = <X, Y> / (||X||*||Y||)
    pairwise_dist = pairwise_distances(idf_title_features,idf_title_features[doc_id])

    # np.argsort retournera des index des plus petites distances
    indices = np.argsort(pairwise_dist.flatten())[0:num_results]
    #pdists stokera les petites distances
    pdists  = np.sort(pairwise_dist.flatten())[0:num_results]

    #data frame indices des 9 plus petites distances
    df_indices = list(data.index[indices])

    for i in range(0,len(indices)):
        get_result(indices[i],data['title'].loc[df_indices[0]], data['title'].loc[df_indices[i]], data['medium_image_url'].loc[df_indices[i]], 'idf')
        print('ASIN :',data['asin'].loc[df_indices[i]])
        print('Brand :',data['brand'].loc[df_indices[i]])
        print ('euclidean distance from the given image :', pdists[i])
        print('='*125)

idf_model(12566,20)
# Dans la heat map affichée, chaque valeur représente les valeurs idf du mot d'étiquette
# la couleur représente l'intersection avec le titre d'entrées
```
